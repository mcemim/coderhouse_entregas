 #### 1. Com este arquivo, construir um objeto de Serie cronológica com o índice igual ao ano dos jogos olímpicos

```python
import pandas as pd
import missingno as msno 
```

```python
# Importando o arquivo
df = pd.read_csv("dbs/Summer-Olympic-medals-1976-to-2008.csv", sep=',', encoding='latin-1')
df.head(2)
```

```python
msno.matrix(df) 
```

```python
#Setando o ano como index
df.set_index("Year", inplace = True)
df.head(2)
```

#### 2. Separar os nomes próprios de sobrenomes em duas colunas distintas usando vírgula como separador.

```python
names = df['Athlete'].str.split(",",  expand = True)    # Cria série separando o nome pela vírgula
df['Name'] = names[1]                                   # Define o segundo item como sendo o nome
df['Lastname'] = names[0]                               # Define o primeiro item como sendo o sobrenome
df.head(2)
```

#### 3. Obter medidas resumidas do conjunto de dados. Qual país ganhou mais medalhas? 

```python
df.describe()

# USA foi o país que mais ganhou medalhas, com 1992 medalhas no total.
```

#### 4. Construir uma tabela mostrando quantas medalhas foram ganhas pelos homens no total em cada ano em que o evento foi realizado.

```python
pd.Series(df[df['Gender'] == 'Men'].groupby(by = 'Year')['Medal'].count())
```
